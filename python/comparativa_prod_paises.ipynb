vamos a hacer un diagrama de  torta desplazada comparativo de la produccion de la principal fuente de Energia en el a√±o 1965 y 2021 entre las zonas de Africa, Asia, europa y Sur america


```python
import matplotlib.pyplot as plt
import pandas as pd

datos = pd.read_csv("produccion_energia_ren.csv")

#Despues de tener la base de datos leida  empezamos a hacer estadisticos con cada zona

Asia = datos[datos['Entity']=='Asia']
Africa =  datos[datos['Entity']=='Africa']
Europa =  datos[datos['Entity']=='Europe']
Suramerica = datos[datos['Entity']=='South America']


#Calculamos el lapso de tiempo
a√±os = Africa['Year']


#Vamos a hacer un gr√°fico de lineas

Asia1 = Asia['Electricity from hydro (TWh)']
Africa1 = Africa['Electricity from hydro (TWh)']
Europa1 = Europa['Electricity from hydro (TWh)']
Suramerica1 = Suramerica['Electricity from hydro (TWh)']




plt.figure(figsize=(15,6))
plt.plot(a√±os,Asia1, label='Asia')
plt.plot(a√±os,Africa1, label='africa')
plt.plot(a√±os,Europa1, label='Europa')
plt.plot(a√±os,Suramerica1, label='Suramerica')

#Etiquetas de la grafica
plt.title('Comparativa de  la Produccion de Energ√≠a Hidroelectrica',fontsize=14)
plt.xlabel('A√±os',fontsize=14)
plt.ylabel('Energia (TWh)',fontsize=14)


plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('comparativa_produccion_hidroelectrica.png')
plt.show()
```

**Ahora vamos a crear un diagrama de barras apiladas con las fuentes principales de energ√≠a renovable de cada pa√≠s 

```python
import matplotlib.pyplot as plt
import pandas as pd

datos = pd.read_csv("produccion_energia_ren.csv")

#Despues de tener la base de datos leida  empezamos a hacer estadisticos con cada zona

Asia = datos[datos['Entity']=='Asia']
Africa =  datos[datos['Entity']=='Africa']
Europa =  datos[datos['Entity']=='Europe']
Suramerica = datos[datos['Entity']=='South America']


#Calculamos el lapso de tiempo
a√±os = Africa['Year']


#Vamos a hacer un gr√°fico de lineas

Asia1 = Asia['Electricity from hydro (TWh)']
Africa1 = Africa['Electricity from hydro (TWh)']
Europa1 = Europa['Electricity from hydro (TWh)']
Suramerica1 = Suramerica['Electricity from hydro (TWh)']




#Vamos a crear el diagrama de barra apiladas
plt.figure(figsize=(15,6))
plt.stackplot(a√±os,Asia1,Africa1,Europa1,Suramerica1,labels=['Asia','Africa','Europa','Suramerica'],
              colors=["#f00c0c","#7c0a90","#1965b2",'#f4f80b'])


#A√±adimos titulos y etiquetas 

plt.title('Comparativa produccion Energ√≠a Hidroelectrica entre diferentes zonas',fontsize=14)
plt.xlabel('A√±os',fontsize=14)
plt.ylabel('Consumo (TWh)',fontsize=14)

#leyenda
plt.legend()

#Mostrar gr√°fica
plt.grid(True)
plt.tight_layout()
plt.savefig('comparativa_produccion_hidroelectrica_diragrama_areas.png')
plt.show()
```
